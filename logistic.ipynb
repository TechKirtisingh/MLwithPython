```python
# in logistic regression it support binary classification mainly..
# but here it is given 3 categories..
# as we work on only 2 category(binary classification) so we rem

df=df[df['species'] != 'setosa'] # here we remove setosa value 
df.head()

# df[] => use to filter the data (whose value is not setosa)
```

```python
df['species']=df['species'].map({'versicolor ':0,'virginica':1}) # map function is used to convert string into integer
```

```python
# Split dataset into independent and dependent features

X=df.iloc[:,:-1] # except last column choose all the column
Y=df.iloc[:,-1] # choose only last column
```
