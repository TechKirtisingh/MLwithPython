# Numpy

numerical python is used for:
. Fast mathemaatical operations
. Working arrays
. Handling large datasets efficiently

Why NumPy ?

. Faster than python lists
.less memory
. support vectorized


```python
# NumPy and Pandas

import numpy as np
```

```python
a = np.array([1, 2, 8, 40, 56])
print(a)
```

```python
# 2D array
b= np.array([[1,2,3] , [4,5,6]])
print(b)
```

```python
c= a + 5 # element-wise addition in the above array
print(c)

d= a*3 # element-wise multiplication in the above array
print(d)    
```

```python
print("Mean" , np.mean(a)) # mean of the array
print("Median" , np.median(a)) # median of the array

print("Sum : ", np.sum(a)) # sum of the array
print("Maximum :" , np.max(a)) # maximum value in the array
print("Minimum :" , np.min(a)) # minimum value in the array
```

```python
# indexing
arr = np.array([10, 20, 30, 40 , 50])
print(arr[3])

# slicing
print(arr[0:3]) # slicing from index 0 to 2 (3 is exclusive)  
print(arr[2:]) # slicing from index 2 to the end of the array
print(arr[:4]) # slicing from the beginning of the array to index 3 (4
print(arr[-3:-1]) # slicing the last 3 elements of the array
```

# PANDAS

Pandas is used for:
. Data manipulation and analysis
. Handling structured data (tables)
. Working with CSV , Excel files


```python
import pandas as pd
```

```python
data ={
    'Name': ['Aman' , 'Riya' , 'Kriti' , 'Rahul'],
    'Marks': [85, 92, 78, 90],
    'Age': [18,17,19,18]
}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
print(df.shape) # shape of the DataFrame (number of rows, number of columns)

print(df.columns) # column names of the DataFram

print(40*"__")

print(df.info()) # information about the DataFrame (data types, non-null values, memory usage)
```

```python
print(df[["Name", "Age"]]) # accessing specific columns of the DataFrame
```

```python
filtered = df[df["Marks"]>80] # filtering rows based on a condition
print(filtered)
```

```python
df["Passed"] = df["Marks"] > 80 # adding a new column to the DataFrame based on a condition
print(df)
```

```python
df_sort = df.sort_values(by="Marks", ascending=False) # sorting the DataFrame by the "Marks" column in descending order
print(df_sort)
```
