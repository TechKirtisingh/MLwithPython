```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

     #####   UNIVARIATE ANALYSIS        
     
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
# print(df.head())

df_setosa=df.loc[df['species']== 'setosa'] # selecting rows where species is setosa

df_verginica=df.loc[df['species']== 'virginica'] # selecting rows where species is virginica
df_versicolor=df.loc[df['species']== 'sversicolor'] # selecting rows where species is versicolor

plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']) , 'o')
plt.plot(df_verginica['sepal_length'],np.zeros_like(df_verginica['sepal_length']) ,'o')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']), 'o')
plt.show()
print(plt.show)
```
