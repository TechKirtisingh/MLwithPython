```python
import matplotlib.pyplot as plt 
# %matplotlib inline ## if we not written this we have to use plt.show....
```

```python
import numpy as np

x=np.arange(0,10)
y=np.arange(11,21)
print(x)
print(y)

a=np.arange(40,50)
b=np.arange(50,60)
print(a)
print(b)
```

```python
print(plt.scatter(x,y,c='b',))## g:green, ploting x and y value with green color dot
plt.xlabel('X axis') # naming the coordinate
plt.ylabel('Y axis')
plt.title('Graph in 2D')
print(plt.savefig('Test.png'))
print(plt.show())
```

```python
print(plt.plot(x, y,'ro--')) ## r-- means redcolor with dotted line and circle
print(plt.show())
print(plt.plot(x, y, 'r*', linestyle='dashed', linewidth=2, markersize=12))
print(plt.xlabel('X-Asis'))
print(plt.ylabel('Y-Asis'))
print(plt.title('Quadratic x2'))
print(plt.show())
```

```python
##$$$$$ CREATING SUBPLOTS

## (nrow,ncol,positon) define..

print(plt.subplot(2,2,1))
print(plt.plot(x,y,'b--'))
print(plt.subplot(2,2,2))
print(plt.plot(x,y,'g*--'))
print(plt.subplot(2,2,3))
print(plt.plot(x,y,'r^'))
print(plt.subplot(2,2,4))
print(plt.plot(x,y,'o'))
print(plt.show())
```

```python
x=np.arange(1,11)
y=3*x^2+5
plt.title("Demo")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x, y, 'r--', label='y=3x+5')
print(plt.show())
```

```python
# Computing the x and y coordinates for points on a sine wave

x= np.arange(0,4*np.pi,0.1) # x values from 0 to 4π with gap of 0.1
y=np.sin(x) 
plt.title("Sine Wave")
print(plt.plot(x,y ))
print(plt.show())
```

```python
## subplot()
# computing the x and y coordinates for points on a sine wave and cosine wave

x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
```

```python
#set up a subplot grid with 2 rows and 1 column
plt.subplot(2,1,1) 
# make the first plot
plt.title("Sine Wave")
print(plt.plot(x, y_sin , 'r--'))

# set up the second subplot
plt.subplot(2,1,2) # it describe the positon of the graph display
plt.title("Cosine Wave")
print(plt.plot(x, y_cos, 'og'))

# show the plots
print(plt.show())
```

```python
##  BAR GRAPH

x=[2,4,8]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y,align='center')
plt.bar(x2,y2,color='r')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
print(plt.show())
```

```python
## HISTOGRAM....

a=np.array([22,87,67,56,44,21,13,11,3,4,78,8,78])
plt.hist(a,bins=20, color='g', edgecolor='black') # bins: number of bars
plt.title('Histogram')
print(plt.show())
```

```python
##  BOX PLOT USING MATPLOTLIB

data= [np.random.normal(0, std, 100) for std in range(1, 4)]
# graph is in percentile form
# rectangle box plot
print(plt.boxplot(data, vert=True, patch_artist=True)) # vert: vertical box plot
#patch_artist: fill the box with color
print(plt.show())
```

```python
## PIE CHART

labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [215, 130, 245, 210]  
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0.01, 0)  # explode : how much far the slice will be from the center

# Plotting the pie chart
##====> pie consist of four parameter that is written above


plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
print(plt.show())
```
